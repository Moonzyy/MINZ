```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x,y = np.meshgrid(np.arange(-10, 10, 0.1), np.arange(-10, 10, 0.1))
```

```python
F1 = np.cosh(x) * np.sin(y) - y
```

```python
F2 = np.sinh(x) * np.cos(y) - x
```

```python
fig, ax = plt.subplots()
ax.contour(x,y,F1,levels = 10,colors = 'r')
ax.contour(x,y,F2, levels = 10,colors = 'b')
fig.set_figwidth(10)   
fig.set_figheight(10) 
ax.grid()
plt.show()
```

```python
from math import cosh, sin, sinh, cos
for i in np.arange(10):
    xn = xy[0]
    yn = xy[1]
    W = np.array([[sin(yn)*sinh(xn), cos(yn)*cosh(xn) - 1],
       [cos(yn)*cosh(xn) - 1, -sin(yn)*sinh(xn)]])
    w = np.linalg.inv(W)
    f = np.array([cosh(xn) * sin(yn) - yn, sinh(xn) * cos(yn) - xn])
    x1 = float("{0:.16f}".format(xy[0]))
    y1 = float("{0:.16f}".format(xy[1]))
    f1 = float("{0:.16f}".format(f[0]))
    f2 = float("{0:.16f}".format(f[1]))  
    print(x1, y1,f1,f2,i)
    xy = xy - np.dot(w,f)
```

# Лабораторная работа №3

```python
from scipy import integrate
from math import exp,sqrt
from sympy import *
```

```python
def lab(L):
    y = lambda t: exp(-L * t) * sqrt(1 + t ** 2)
    result = integrate.quad(y, 0, 10)
    print(result)
    for N in range(0,8,2):
        ans = 0
        x = Symbol('x')
        f = exp(-L * x) * sqrt(1 + x ** 2)
        for k in range(N + 1):
            ans += float(f.diff(x,k).subs(x,0) / (L ** (k + 1)))
        print('N = ', N, ':', ans, 'погрешность:', result[0] -ans)
        
```

```python
lab(1)
lab(5)
lab(7)
lab(10)
lab(15)
```
